```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv('personal_transactions.csv')
raw_df
```

### Changing the name of the column to remove spaces between each word

```python
raw_df.rename(columns={'Transaction Type':'Transaction_Type', "Account Name":'Account_Name'}, inplace= True)
raw_df
```

### Adding Seasons column

```python
seasons_added_df = raw_df.copy()
for row in range(len(raw_df)):
    #Look at data date format and see if it is '1' or '01'
    if ('4') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Spring'
    if ('5') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Spring'
    if ('6') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Spring'
        
    if ('7') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Summer'
    if ('8') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Summer'
    if ('9') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Summer'
        
    if ('1') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Winter'
    if ('2') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Winter'
    if ('3') in raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Winter'
        
    if ('10') == raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Fall'
    if ('11') == raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Fall'
    if ('12') == raw_df['Date'][row][:2]:
        seasons_added_df.at[row,'Seasons'] = 'Fall'
seasons_added_df
```

### Dividing the dataset into year 2018 and 2019

```python
year_2018_df = seasons_added_df.copy()
for rows in range(len(year_2018_df)):
    if '2019' in (year_2018_df['Date'][rows]):
        year_2018_df.drop(rows, inplace = True)
year_2018_df
```

```python
year_2019_df = seasons_added_df.copy()
for rows in range(len(year_2019_df)):
    if '2018' in (year_2019_df['Date'][rows]):
        year_2019_df.drop(rows, inplace = True)
        
year_2019_df.reset_index(inplace = True)
year_2019_df.drop(columns= ['index'], inplace = True)
year_2019_df
year_2019_df
```

### Createing its own paycheck dataframe

find the indicies where `Category` is equal to Paycheck

```python
paycheck_condition = seasons_added_df['Category']=='Paycheck'
paycheck_condition
```

```python
s_added_index = seasons_added_df.index
s_added_index
```

```python
paycheck_indicies = s_added_index[paycheck_condition]
paycheck_indicies
```

```python
paycheck = paycheck_indicies.tolist()
paycheck
```

### Creating Paycheck only dataframe

```python
paycheck_df = pd.DataFrame()
for paycheck_index in paycheck:
    paycheck_df =  paycheck_df.append(seasons_added_df.loc[paycheck_index])

#Resetting index
paycheck_df.reset_index(inplace = True)
paycheck_df.drop(columns= ['index'], inplace = True)
paycheck_df
```

### Creating its own payment dataframe
- Same process as creating paycheck dataframe

```python
payment_condition = seasons_added_df["Category"]=="Credit Card Payment"
s_added_index = seasons_added_df.index
payment_indicies = s_added_index[payment_condition]
payment_indicies = payment_indicies.tolist()
payment_indicies
```

```python
credit_payment_df = pd.DataFrame()
for payment_index in payment_indicies:
    credit_payment_df =  credit_payment_df.append(seasons_added_df.loc[payment_index])

#Resetting index
credit_payment_df.reset_index(inplace = True)
credit_payment_df.drop(columns= ['index'], inplace = True)
credit_payment_df
```

### Removing `Credit Card Payment` and `Paycheck` from `Category` column

```python
spending_transaction_df = seasons_added_df.copy()
for payment in payment_indicies:
    spending_transaction_df = spending_transaction_df.drop([payment])
for deposit in paycheck_indicies:
    spending_transaction_df = spending_transaction_df.drop([deposit])
spending_transaction_df = spending_transaction_df.reset_index(drop=True)
spending_transaction_df
```
